# Types of Brands and number of cars in each brand

```python
fig, ax = plt.subplots()
sns.countplot(x = "Brand", data = df, ax = ax)
for i in ax.containers:
    ax.bar_label(i)
plt.xlabel("Brands")
plt.ylabel("Number of cars")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.title('Numbre of Cars in each Brand')
```

# Cars with number of owners

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.countplot(x ='Owner_Type', data = df,ax=ax)
for i in ax.containers:
    ax.bar_label(i)
plt.title('Cars with no of previous owner')
```

# Cars with the total kilometers driven

```python
fig, ax = plt.subplots(figsize = (30,10))

plt.bar(df["Model"], df["Kilometers_Driven"])
for i in ax.containers:
    ax.bar_label(i)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.title("Cars with their kilometers driven")
plt.show()
```

# Crs with total number of seats

```python
fig, ax = plt.subplots(figsize = (30,5))

plt.bar(df["Model"], df["Seats"])
for i in ax.containers:
    ax.bar_label(i)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.title("Cars with number of seats")
plt.show()
```

# Cars wtih fuel type

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.countplot(x ='Fuel_Type', data = df,ax=ax)
for i in ax.containers:
    ax.bar_label(i)
plt.title('Cars with petrol or desiel fuel')
```

# Transmission difference between cars

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.countplot(x ='Transmission', data = df,ax=ax)
for i in ax.containers:
    ax.bar_label(i)
plt.title('Cars with transmission')
```

```python
sns.set_theme(style = "darkgrid")

g = sns.jointplot(x  = "Engine", y = "Power", data = df, kind = "reg")
plt.title("Engine and Power graph")
```

# Model with their price

```python
fig, ax = plt.subplots(figsize = (30,5))

plt.bar(df["Model"], df["Price"])
for i in ax.containers:
    ax.bar_label(i)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.title("Price for each model")
plt.show()
```
